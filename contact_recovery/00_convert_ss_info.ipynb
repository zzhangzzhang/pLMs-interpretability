```python
import json

def parse_input(input_content):
    lines = input_content.strip().split('\n')
    data_dict = {}
    for line in lines:
        structure, filename = line.rsplit(' ', 1)
        data_dict[filename] = structure
    return data_dict

def save_to_json(data, filename):
    with open(filename, 'w') as f:
        json.dump(data, f, indent=4)
        
with open('../data/output.result', 'r') as f:
    content = f.read()

parsed_data = parse_input(content)
save_to_json(parsed_data, '../data/filtered_ss.json')
```
